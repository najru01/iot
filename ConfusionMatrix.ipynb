```python
import numpy as np
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
actual= np.array(['mango','mango','mango','Not mango','mango','Not mango','mango','mango','Not mango','Not mango'])
```

```python
predicted = np.array(
    ['mango','Not mango','mango','Not mango','mango','mango','mango','mango','Not mango','Not mango'])
```

```python
cm = confusion_matrix(actual,predicted)
```

```python
sns.heatmap(cm,
            annot=True,
            xticklabels=['mango','Not mango'],
            yticklabels=['mango','Not mango'], cmap="Blues")
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=15)
plt.show()
```
