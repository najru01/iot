```python
from sklearn.datasets import load_iris
```

```python
datairis=load_iris()
```

```python
datairis.feature_names
```

```python
df=pd.DataFrame(datairis.data,columns=datairis.feature_names)
```

```python
df.head()
```

```python
# Appending target column 
```

```python
df['target']=datairis.target
```

```python
df
```

```python
# 0--> Setosa 1-->Versicolor 2-->Virginica 
```

```python
# we can visualize each class of target (0,1,2)
```

```python
df[df.target==1]
```

```python
# we can add one more column into df as "fname"
```

```python
# by using Lmbda method
```

```python
df['fname']=df.target.apply(lambda x : datairis.target_names[x])
```

```python
# X-->input Y-->output
```

```python
X=df.drop(['target','fname'],axis='columns')
```

```python
X
```

```python
y=df.target
```

```python
y
```

```python
# Splitting data into train and test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# total rows =150
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_test
```

```python
# Importing SVM 
```

```python
from sklearn.svm import SVC
```

```python
# here SVC--> support vector Classifier
```

```python
svmodel=SVC(kernel='linear')
```

```python
svmodel.fit(X_train,y_train)
```

```python
svmodel.predict([[5.7,2.6,3.5,1.0]])
```

```python
svmodel.predict([[7.1,3.0,5.9,2.1]])
```

```python
y_predict=svmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_predict,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_predict,y_test)
```
